```python
from ultralytics import YOLO
import os
from IPython.display import display,Image
from IPython import display
display.clear_output()
!yolo checks
```

```python
model = YOLO('yolov8n.pt')  # Load a predefined model
res = model.train(data="/content/drive/MyDrive/data.yaml", epochs=150)
```

```python
model.save('/content/drive/MyDrive/yolov8n_trained.pt')
```

```python
# Directory containing images to predict
image_directory = '/content/sample'

# Run inference on all images in the directory
results = model.predict(source=image_directory, save=True, save_txt=True, save_conf=True)
```
